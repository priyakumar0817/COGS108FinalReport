## Background and Prior Work

The unlawful killing of one person by another, also known as violent deaths or homicides, has been an understandable concern in America throughout history. We define intentional homicides as individuals or small groups who commit deaths inflicted as a result of domestic disputes, interpersonal violence, violence conflicts over land resources, and intergang violence over turf or control **[1]**. In recent years, there has been a 30 percent increase in U.S. homicide rates in 2020, the biggest national homicide crime rate in over a century **[2]**. The pandemic surge in violence continued into 2021 with a rise in homicides by 4.3 percent since 2020 **[3]**. There are many socioeconomical and sociodemographic factors to be considered in murder rates, including the political differences among Americans that are deeply rooted throughout history.

**The two major political parties in America are Republican and Democrat**. Both parties have polarizing views, with Democratic states that follow liberal ideologies, democracy, maintaining world peace, championing human rights, and seek change from the traditional values and policies of government **[4]**. Republican states oppose this viewpoint, and seek to maintain a conservative ideology, with an emphasis on traditional forms of power and control over human rights and economic development.

As a result of these opposing viewpoints on American governance, prior work has suggested that the major political party of a given state has been significantly correlated with the changing trends of homicide rates over time. Homicide rates were higher for states with higher votes of the Republican running candidate than in states who voted for the Democratic candidates in the years 1900-2020 **[5]**. Research suggests that the polarization between the parties and their policies have strong effects on domestic issues, as the policies of Democratic administrations decrease socioeconomic distress that support the psychology and sensibility of indivuals, ultimately reducing levels of murder rates, as opposed to Republican policies that consequently result in an increase of violent death rates **[4]**.

While prior work on this relationship seems to suggest that changes in homicides rates are more likely due to differences in socioeconomic policies between parties that lead to different psychosocial effects among individuals, party effect is large enough to overpower other contributors that make political viewpoint between states an important level of consideration **[4]**. We are interested in finding if there is a difference in the distribution of homicide rates between Republican and Democratic states.

## Data Cleaning

### **Population Data Cleaning:**  

We slice our population data so that the Year variable only contains the range **1980-2010**. We also slice by the Geography Type so that it only contains states.

```python
# Filter by relevant Year range of 1980-2010
population = population.loc[(population['Year'] >= 1980) & (population['Year'] < 2010)]
# Only include State from the Geography Type column
population = population.loc[population['Geography Type'] == 'State']
population
```

Obtain relevant columns from population dataset, and **do not include data from Puerto Rico** as it is not included as part of the 50 U.S. States, so will not be considered.

```python
# Obtain relevant columns needed, and reset index of population dataframe
population = population[['Name', 'Year', 'Resident Population']].reset_index(drop=True)
# Remove Puerto Rico from Population data
population = population.loc[(population['Name'] != 'Puerto Rico')]
# Print all unique values of 'Name', representing States 
population['Name'].unique()
```

Since the Resident Population column was read in as a string with commas, we **standardize** this data by removing the commas and change the data type in each column to an integer. 

```python
population['Resident Population'] = population['Resident Population'].apply(lambda x: int(str(x).replace(',', '')))
population
```

**Verify** that there are no null data in our population dataset. Null count must equal 0. 

```python
population.isnull().sum()
```

### **Election Data Cleaning:**
First, we need to match the years from our election dataset to our homicide dataset. Slice the election dataset to include only the years between **1980-2010**.

```python
elections = elections.loc[(elections['year'] >= 1980) & (elections['year'] < 2010)]
elections
```

Obtain the relevant columns in the election dataset.

```python
elections = elections.drop(['notes', 'party_detailed', 'state_fips', 'state_cen', 'state_ic', 'writein', 'office', 'version'], axis=1)
elections
```

Obtain the two major parties of interest (**Democratic and Republican Parties**). 

```python
elections = elections.loc[(elections['party_simplified'] == 'REPUBLICAN') | (elections['party_simplified'] == 'DEMOCRAT')]
elections
```

**Verify** that there is no null data in our elections dataset. Null count must equal 0. 

```python
elections.isnull().sum()
```

Add a **Start Year** column to represent the start year of each decade.

```python
# Change string data to an integer number, and round each Year to its respective Decade 
elections['Start Year'] = elections['year'].apply(lambda x: int(str(x)[:-1] + '0'))
elections.head()
```

Create two separate dataframes from the current elections dataframe, one for Republicans and another for Democrats.

```python
# Two dataframes per major party
dems = elections.loc[elections['party_simplified'] == 'DEMOCRAT']
reps = elections.loc[elections['party_simplified'] == 'REPUBLICAN']
# Print Democratic dataframe 
dems
```

**Group each dataframe** by the decade and state, and combine all the vote counts for both dataframes. 

```python
# Display candidate and total votes from each party, and group them by the year and state
# Reset index to reorganize the dataframes
dems = dems.groupby(['Start Year', 'state'])['candidatevotes', 'totalvotes'].sum().reset_index()
reps = reps.groupby(['Start Year', 'state'])['candidatevotes', 'totalvotes'].sum().reset_index()
dems
```

Calculate the **proportion** of people who voted Democrat or Republican within each decade for each state.

```python

# Obtain the proportion for each major party by dividing the candidate votes by the 
# total votes for each state
dems['proportion'] = dems['candidatevotes']/dems['totalvotes']
reps['proportion'] = reps['candidatevotes']/reps['totalvotes']
dems
```

Since the state names are fully capitalized in the election dataframes, we define a function that **standardizes** the state names to only the first letter in each word.

Only keep the first letter capitalized (except for the 'of' in 'District of Columbia') to match the state values in our homicide dataset. **We apply this function to the state column of the Democrat and Republican datasets**.

```python
def standardize_state(x):
    x = x.split(' ')
    s = ''
    for i in x:
        if i == 'OF':
            s += i.lower() + ' '
        else: 
            s += i[0] + i.lower()[1:] + ' '
    return s[:-1]
    
dems['state'] = dems['state'].apply(standardize_state)
reps['state'] = reps['state'].apply(standardize_state)
dems
```

### **Homicide Data Cleaning**

We obtain the state homicide data per *decade*.We select the relevant columns to store in a dataframe state_homicide. The lambda function rounds the Year columns down to the decade and converts it to an integer. We only include data from **1980-2010**. 

```python
state_homicide = homicides[['Year', 'State']]
state_homicide['Year'] = state_homicide['Year'].apply(lambda x: int(str(x)[:-1] + '0')) 
state_homicide = state_homicide.loc[state_homicide['Year'] < 2010]
```

Get the homicide count for each state per decade. Print California data to **verify the years are every ten years**.

```python
# Group out dataframe by year and for a given state, and reset the dataframe's index
state_homicide = state_homicide.groupby(['Year'])['State'].value_counts()
state_homicide = state_homicide.to_frame().rename({"State": "Homicide Count"}, axis=1).reset_index()
# Obtain relevant columns relating to the year, state, and homicide counts
print(state_homicide.head())
state_homicide[state_homicide['State'] == 'California']
```

**Verify** that there are no null data in our state homicide dataset. Null count must equal 0. 

## **Merging Datasets**
Rhode Island in the state_homicide dataset is spelled as "Rhodes Island." Change this to Rhode Island, and merge the state homicide count with the state population data.

```python
# Clean naming of "Rhodes Island"
state_homicide['State'] = state_homicide['State'].apply(lambda x: 'Rhode Island' if x == 'Rhodes Island' else x)
# Merge state homicide with population dataset and store merge results back into variable state_homicide
state_homicide = state_homicide.merge(population, left_on=['Year', 'State'], right_on=['Year', 'Name'], how='outer')
state_homicide
```

There exists columns of the same state name, so drop these duplicate columns after merging the homicide data with the population data.

```python
state_homicide = state_homicide.drop(['Name'], axis = 1)
state_homicide
```

### **Calculate Homicide Rate**  
Create a new column in the state homicide dataset that calculates the homicide rate from: Homicide Count and Resident Population. 

```python
# Calculate the homicide rate by dividing the state homicide count by resident population
state_homicide['Homicide Rate'] = state_homicide['Homicide Count']/state_homicide['Resident Population']
state_homicide
```

Since our analysis is based on decades, we differentiate the 'Year' column by renaming it 'Start Year,' and increase 'Year' by ten to dictate the 'End Year.'

```python
# Create 'End Year' column by adding 10 to the current year
state_homicide['End Year'] = state_homicide['Year'].apply(lambda x: x + 10)
# Rename 'Year' as the 'Start Year'
state_homicide = state_homicide.rename(columns={'Year' : 'Start Year'})
# Change the order by which variables are presented for better readability
state_homicide = state_homicide[['Start Year', 'End Year', 'State', 'Homicide Count','Resident Population', 'Homicide Rate']]
state_homicide
```

**Merge** the state homicide count dataset with the democratic dataset using the Start Year and State column, and drop the repeated state column. 

```python

state_homicide = state_homicide.merge(dems[['Start Year', 'proportion', 'state']], left_on=['Start Year', 'State'], 
                                      right_on=['Start Year', 'state'])
# Rename proportion column to coincide with Democrat proportion 
state_homicide = state_homicide.rename(columns={'proportion' : 'Democrat Proportion'})
state_homicide = state_homicide.drop(['state'], axis = 1)
state_homicide
```

**Merge** the state homicide count dataset with the republican dataset using the Start Year and State column, and drop the repeated state column.

```python
state_homicide = state_homicide.merge(reps[['Start Year', 'proportion', 'state']], left_on=['Start Year', 'State'], 
                                      right_on=['Start Year', 'state'])
# Rename proportion column to coincide with Republican proportion 
state_homicide = state_homicide.rename(columns={'proportion' : 'Republican Proportion'})
state_homicide = state_homicide.drop(['state'], axis = 1)
state_homicide
```

# Inferential Analysis

In order to answer our question about whether there is a difference in distribution of homicide rate between Republican and Democratic states, we decided to use **t-tests**. In our EDA, we found that homicide rate **decreased** over time, so other than our overall t-test from 1980-2010, we decided to break down each t-test by **decade** so that an imbalance of votes (like during 1980-1990 where the states were overwhelmingly Republican) does not skew the test.


In order to perform t-tests for the distribution of Homicide Rates between Republican and Democratic states, **the following assumptions should be met**:

**1. Independence**: The observations in the distribution of Homicide Rates for Republican States are independent of the observations in the distribution of Homicide Rates for Democratic States.

**2. Normality**: Both distributions for each state Both samples are approximately normally distributed.

**3. Homogeneity of Variances**: Both samples have approximately the same variance.

**4. Random Sampling**: Both samples were obtained using a random sampling method.

We decided to use an **significance level of 0.01** for our t-tests so that if there is a significant difference, it is more likely to actually be significant.
We will provide the t-test results for 1980-2010, followed by a t-test for each decade, and interpretation of the results after all t-tests are performed.

### T-Test between Republicans and Democrats

Separate the distributions of Republicans and Democrats for homicide rates and perform a t-test to find if there is a significant difference between distributions from **1980-2010**.

```python
rep_all = state_homicide[state_homicide['Higher Prop'] == 'Republican']['Homicide Rate']
dem_all = state_homicide[state_homicide['Higher Prop'] == 'Democrat']['Homicide Rate']
```

```python
t_val, p_val = ttest_ind(dem_all,rep_all)
print(t_val,p_val)
if p_val < 0.01:
    print('There is a significant difference in the distributions of Republican and Democrat state homicide rates from 1980-2010!')
else:
    print('There is NOT a significant difference in the distributions of Republican and Democrat state homicide rates from 1980-2010!')
```

## Regression

After seeing how our t-test's p-value improved by removing swing states, we hypothesized that **maybe there is a pattern in homicide rates based on how far leaning they were**. We performed a linear regression on the difference between the Republican voting proportion and Democrat voting Proprtion to predict the homicide rate. 

```python
outcome, predictors = patsy.dmatrices('Q("Homicide Rate") ~ Q("Difference in Proportion")', data=state_homicide)
model = sm.OLS(outcome, predictors)
res1 = model.fit()
print(res1.summary())
```

We can see that the difference between the Republican voting proportion and Democrat voting Proprtion is **not a significant variable** to predict homicide rate. **A positive coefficient** means that as difference in proportion increases, homicide rate increases a little.

```python
sns.lmplot(x='Difference in Proportion',y='Homicide Rate',data=state_homicide,fit_reg=True,height=8,aspect=1.8)
ax = plt.gca()
ax.set_title("Linear Regression Model of Difference in Proportion to Homicide Rate");
```

Looking at our data we can see a pattern that the majority of the points on the far-leaning states are **below the line** while in the middle, our swing states are mostly above the line. This may mean that a line may not be able to fit this data, instead let's look at a **quadratic function**, which can fit the pattern with a parabola.

```python
outcome, predictors = patsy.dmatrices('Q("Homicide Rate") ~ Q("Difference in Proportion") + I(Q("Difference in Proportion")**2)',
                                      data=state_homicide)
model = sm.OLS(outcome, predictors)
res2 = model.fit()
print(res2.summary())
```

Using a quadratic function, we find that our Difference in Proportion squared is a significant variable with **p-value < 0.01**, while the original Difference in Proportion is much closer to significant than before. 

```python
sns.regplot(x='Difference in Proportion',
            y='Homicide Rate',
            data=state_homicide,order=2).set_title(
    "Regression of Quadratic Model of Difference in Proportion to Homicide Rate");
```

We can see that our proposed parabola better fits the function, with Homicide Rates peaking with swing states where tensions are higher, which **affirms our secondary hypothesis about swing states**. We will discuss our results from the entire analysis in our conclusion and discussion section.
